```python
class NativeDictionary:
    def __init__(self, sz):
        self.size = sz
        self.slots = [None] * self.size
        self.values = [None] * self.size

    def hash_fun(self, key):
        k = len(key)%self.size
        return k

    def is_key(self, key):
        k = self.hash_fun(key)
        
        if self.slots[k] is None:
            return False
        
        if key == self.slots[k]:
            return True
        elif key in self.slots[k]:
            return True
        return False

    def put(self, key, value):
        k = self.hash_fun (key)
        
        if self.slots[k] is None:
            self.slots[k] = key
            self.values[k] = value
            return
        
        if self.slots[k] == key:
            self.values[k] = value
            return
        
        if key in self.slots[k]:
            self.values[k][self.slots[k].index(key)] = value
            return           
            
        if type(self.slots[k]) == str:
            key_to_append = self.slots[k]
            value_to_append = self.values[k]
            self.slots[k] = []
            self.slots[k].append(key_to_append)
            self.slots[k].append(key)
            self.values[k] = []
            self.values[k].append(value_to_append)
            self.values[k].append(value)
                    
        elif type (self.slots[k]) == list:
            self.slots[k].append(key)
            self.values[k].append(value)

    def get(self, key):
        k = self.hash_fun(key)
        if self.is_key(key):
            if self.slots[k] == key:
                return self.values[k]
            return self.values[k][self.slots[k].index(key)]
        return None
```

```python
print ('Тест put добавление нового значения')
dict_to_test = NativeDictionary(19)
dict_to_test.put('Яблоки', 754)
print ('Наличие ключа в словаре', dict_to_test.is_key('Яблоки'), 'Равенство значения ключа', dict_to_test.get('Яблоки')== 754)
```

```python
print ('Тест put добавление нового значения в тот же слот')
dict_to_test = NativeDictionary(19)
dict_to_test.put('Яблоки', 754)
dict_to_test.put('Бананы', 132)
dict_to_test.put('Томаты', 666)
print ('Наличие ключа Яблоки в словаре', dict_to_test.is_key('Яблоки'), 'Равенство значения ключа', dict_to_test.get('Яблоки')== 754)
print ('Наличие ключа Бананы в словаре', dict_to_test.is_key('Бананы'), 'Равенство значения ключа', dict_to_test.get('Бананы')== 132)
print ('Наличие ключа Томаты в словаре', dict_to_test.is_key('Томаты'), 'Равенство значения ключа', dict_to_test.get('Томаты')== 666)
```
