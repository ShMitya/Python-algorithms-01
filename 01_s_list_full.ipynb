```python
def sum_two (s_list1, s_list2):
    if s_list1.len() == s_list2.len():
        s_list_sum = LinkedList()
        node1 = s_list1.head
        node2 = s_list2.head
        while node1 and node2 is not None:
            s_list_sum.add_in_tail (Node(node1.value + node2.value))
            node1 = node1.next
            node2 = node2.next
    return s_list_sum
```
